```python
DF=pd.read_csv(r'/content/drive/MyDrive/HR_Attrition_cleaned.csv')
DF.head()
```

```python
DF.info()
```

```python
DF=DF.drop(columns=['Phone'])
```

```python
DF['Attrition_Int']=pd.get_dummies(DF['Attrition'],drop_first=True).astype(int)
```

```python
DF['Hire_Date']=pd.to_datetime(DF['Hire_Date'])
```

```python
DF['Year']=DF['Hire_Date'].dt.year
DF['Month']=DF['Hire_Date'].dt.month
DF['Day']=DF['Hire_Date'].dt.day
```

```python
#DF=DF.drop(columns=['Attrition'])
#DF=DF.drop(columns=['Hire_Date'])
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
```

```python
categorical_feature= ['Department','Job_Title','Education']
num_feature= ['Job_Satisfaction','Year','Month','Day','Salary']

preprocessor= ColumnTransformer(transformers=[
    ('cat',OneHotEncoder(handle_unknown='ignore',sparse_output=False),categorical_feature),
    ('num',StandardScaler(),num_feature)
])
```

```python
model= Pipeline([('preprocess',preprocessor),('model',LogisticRegression(max_iter=1000,class_weight="balanced"))])

X= DF.drop(columns=['Attrition_Int','Attrition','Hire_Date'])
y= DF['Attrition_Int']

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=190)

model.fit(X_train,y_train)
```

```python

from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
y_pred=model.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
acc= accuracy_score(y_test,y_pred)
print(acc)
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.cluster import KMeans
XTrain= DF.drop(columns='Attrition_Int')
X_process= preprocessor.fit_transform(XTrain)
inertia=[]
for k in range(1,11):
  kmeans= KMeans(n_clusters=k,random_state=42,n_init=10)
  kmeans.fit(X_process)
  inertia.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(8,5))
plt.plot(range(1,11),inertia,marker='o',linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans= KMeans(n_clusters=4,random_state=42,n_init=10)
kmeans.fit(X_process)
```

```python
clusters= kmeans.labels_
DF['Clusters']=clusters
```

```python
DF.groupby('Clusters').mean(numeric_only=True)
```

```python
DF.info()
```

```python
DF.groupby('Clusters')['Job_Satisfaction'].value_counts()
```

```python
from sklearn.metrics import silhouette_score
score= silhouette_score(X_process,clusters)
print(score)
```

```python
DF.groupby('Clusters')['Attrition'].value_counts()
```

```python
from sklearn.tree import DecisionTreeClassifier
tree= DecisionTreeClassifier(max_depth=4,min_samples_leaf=3,min_samples_split=3,random_state=5430,class_weight="balanced")
model= Pipeline([('preprocess',preprocessor),('model',tree)])
model.fit(X_train,y_train)
```

```python
pred_y= model.predict(X_test)
print(classification_report(y_test,pred_y))
```

```python
print(confusion_matrix(y_test,pred_y))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=100,max_depth=4,min_samples_leaf=3,min_samples_split=3,random_state=5430,class_weight="balanced",n_jobs=-1)
rf_model= Pipeline([('preprocess',preprocessor),('model',rf)])
rf_model.fit(X_train,y_train)
```

```python
y_pred_rf= rf_model.predict(X_test)
print(classification_report(y_test,y_pred_rf))
```

```python
print(accuracy_score(y_test,y_pred_rf))
```

```python
print(confusion_matrix(y_test,y_pred_rf))
```
